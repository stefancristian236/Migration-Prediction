```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_theme(style='whitegrid')

try:
    df = pd.read_csv('../Data/birds_db.csv')
    print('Data loaded successfully.')
except FileNotFoundError:
    print('Error: birds_db.csv not found. Please upload it.')
    exit()

target_species = 'Collared flycatcher'
bird_data = df[df['species'] == target_species].copy()
print(f'Analyzing: {target_species} ({len(bird_data)} rows)')
```

```python
plt.figure(figsize=(10, 6))

# Plotting with your hue strategy
sns.scatterplot(
    data=bird_data, 
    x='latitude', 
    y='ndvi', 
    hue='phase', 
    palette='viridis', 
    s=100, 
    alpha=0.8,
    edgecolor='w'  # Adds a white border to dots for better separation
)

# Add a reference line for "Healthy Vegetation" (optional but helpful)
plt.axhline(y=0.4, color='r', linestyle='--', alpha=0.5, label='Dense Vegetation Threshold')

plt.title(f'{target_species}: Vegetation Preference by Migration Phase', fontsize=14)
plt.xlabel('Latitude (North)', fontsize=12)
plt.ylabel('Vegetation Health (NDVI)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Life Cycle Phase')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=bird_data, x='latitude', y='ndwi', hue='phase', palette='cool', s=100, alpha=0.8)

plt.title(f'{target_species}: Water Index vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('NDWI (Water Index)')
plt.show()
```

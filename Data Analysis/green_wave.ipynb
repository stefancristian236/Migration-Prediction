```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_theme(style='whitegrid')

try:
    df = pd.read_csv('../Data/birds_db.csv')
    print('Data loaded successfully.')
except FileNotFoundError:
    print('Error: birds_db.csv not found. Please upload it.')
    exit()

target_species = 'Collared flycatcher'
bird_data = df[df['species'] == target_species].copy()
print(f'Analyzing: {target_species} ({len(bird_data)} rows)')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=bird_data, x='latitude', y='ndvi', hue='phase', palette='viridis', s=100, alpha=0.8)
sns.regplot(data=bird_data, x='latitude', y='ndvi', scatter=False, color='gray', line_kws={'linestyle': '--'})

plt.title(f'{target_species}: Vegetation vs Latitude')
plt.xlabel('Latitude (South -> North)')
plt.ylabel('NDVI (Vegetation Index)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=bird_data, x='latitude', y='ndwi', hue='phase', palette='cool', s=100, alpha=0.8)

plt.title(f'{target_species}: Water Index vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('NDWI (Water Index)')
plt.show()
```

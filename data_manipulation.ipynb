```python
import pandas as pd
import numpy as np
import os
```

```python
#define base study birds cluster
d = {
    "species": [
        "European pied flycatcher",
        "Spotted flycatcher",
        "Collared flycatcher",
        "Wood warbler",
        "Willow warbler",
        "Common Redstart",
        "Brambling",
        "Eurasian Siskin",
        "Blackcap",
        "Wood Pigeon",
        'Northern Lapwing',
        'Eurasian Curlew'
        'Black-tailed Godwit',
        'Barn Swallow',
        'Common House Martin',
        'European Golden Plover',
        'White Stork',
        'Common Cuckoo',
        "Common Swift",
        'Eurasian Hoopoe'
    ]
}

#convert into an dataframe
df = pd.DataFrame(data = d)
```

```python
#export dataframe into an .csv file
df.to_csv('dataBase', index = True)
```

```python
data_base = 'dataBase'
df = pd.read_csv(data_base)

forest_dependent = [
    "European pied flycatcher",
    "Spotted flycatcher",
    "Collared flycatcher",
    "Wood warbler",
    "Willow warbler",
    "Common Redstart",
    "Brambling",
    "Eurasian Siskin",
    "Blackcap",
    "Wood Pigeon"
]

non_forest_dependent = [
    "Northern Lapwing",
    "Eurasian Curlew",
    "Black-tailed Godwit",
    "Barn Swallow",
    "Common House Martin",
    "European Golden Plover",
    "White Stork",
    "Common Cuckoo",
    "Common Swift",
    "Eurasian Hoopoe"
]

df["forest_label"] = df["species"].isin(forest_dependent).astype(int)

df['latitude'] = np.nan
df['longitude'] = np.nan
df['ndwi'] = np.nan
df['mndwi'] = np.nan
df['ndbi'] = np.nan
df['ndvi'] = np.nan
df['ndmi'] = np.nan
df['phase'] = np.nan
df['year'] = np.nan
df['month'] = np.nan
df['continet'] = np.nan
df['country'] = np.nan

df.to_csv(data_base, index=False)
print('Update Complete.')
print(df[['species', 'forest_label']].head())
```

```python
#create separate csv files for every kind of bird
base_path = 'species'
forest_path = os.path.join(base_path, 'forest_dependent')
non_forest_path = os.path.join(base_path, 'non_forest_dependent')

os.makedirs(forest_path, exist_ok = True)
os.makedirs(non_forest_path, exist_ok = True)

print('Creating files for forest dependent species')
for species in forest_dependent:
    subset = df[df['species'] == species]

    if not subset.empty:
        filename = f"{species.replace(' ', '_')}.csv"
        full_path = os.path.join(forest_path, filename)
        subset.to_csv(full_path, index = False)
        print(f" -> Saved: {full_path}")
    
print('\nCreating files for non forest dependent species')
for species in non_forest_dependent:
    subset = df[df['species'] == species]

    if not subset.empty:
        filename = f"{species.replace(' ', '_')}.csv"
        full_path = os.path.join(non_forest_path, filename)
        subset.to_csv(full_path, index = False)
        print(f" -> Saved: {full_path}")

print('\n All files created succesfully')
```

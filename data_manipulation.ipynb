```python
import pandas as pd
import numpy as np
```

```python
#define base study birds cluster
d = {
    "species": [
        "European pied flycatcher",
        "Spotted flycatcher",
        "Collared flycatcher",
        "Wood warbler",
        "Willow warbler",
        "Common Redstart",
        "Brambling",
        "Eurasian Siskin",
        "Blackcap",
        "Wood Pigeon",
        'Northern Lapwing',
        'Eurasian Curlew'
        'Black-tailed Godwit',
        'Barn Swallow',
        'Common House Martin',
        'European Golden Plover',
        'White Stork',
        'Common Cuckoo',
        "Common Swift",
        'Eurasian Hoopoe'
    ]
}

#convert into an dataframe
df = pd.DataFrame(data = d)
```

```python
#export dataframe into an .csv file
df.to_csv('dataBase', index = True)
```

```python
data_base = 'dataBase'
df = pd.read_csv(data_base)

forest_dependent = [
    "European pied flycatcher",
    "Spotted flycatcher",
    "Collared flycatcher",
    "Wood warbler",
    "Willow warbler",
    "Common Redstart",
    "Brambling",
    "Eurasian Siskin",
    "Blackcap",
    "Wood Pigeon"
]

non_forest_dependent = [
    "Northern Lapwing",
    "Eurasian Curlew",
    "Black-tailed Godwit",
    "Barn Swallow",
    "Common House Martin",
    "European Golden Plover",
    "White Stork",
    "Common Cuckoo",
    "Common Swift",
    "Eurasian Hoopoe"
]

df["forest_label"] = df["species"].isin(forest_dependent).astype(int)

df['latitude'] = np.nan
df['longitude'] = np.nan
df['ndwi'] = np.nan
df['mndwi'] = np.nan
df['ndbi'] = np.nan
df['ndvi'] = np.nan

df.to_csv(data_base, index=False)
print('Update Complete.')
print(df[['species', 'forest_label']].head())
```
